```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
# create sample data
np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0, 100, 50)

#create a graph 
fig, ax = plt.subplots(figsize=(5,3))
#ax.scatter(x, y)
#ax.bar(x,y)
#ax.barh(x,y)
#ax.plot(x,y)
ax.hist(y)

#add labels and title
```

```python
# create font dictionaries
font1 = {'family':'sans-serif','color':'blue','size':20}
font2 = {'family':'monospace', 'color': 'green', 'size':14}

#create line plot with labels and a title
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)
ax.set_title(' Total growth over time', fontdict = font1, loc ='left')
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)
fig.tight_layout()
```

```python
#change marker of the points
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x, y, marker = '*', color = 'indigo')
```

```python
#change marker of the points and modify the line style
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y, marker = '*', color = 'indigo', linestyle = '--', linewidth = 2)
```

```python
# modify the line plot with custome colors and styles 
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y,
        marker = '*',
        markersize = 12, 
        color = '#7425b9',
        linestyle = '--',
        linewidth = 2, 
        markeredgecolor = "#fa9359",
        markerfacecolor = '#000000'

    )
ax.grid(color = 'blue', linestyle = 'dotted', linewidth = 2) #add grid lines with custom color, style, and width
```
