```python
x1 = np.arange (50)
y1 = np.random.randint(0, 75,50)
x2 = np.array (["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y2 = np.array([110, 180, 240, 99, 220])

```

```python
font1 = {'family':'serif','color':'indigo'}

fig, (ax1, ax2) = plt.subplots(ncols = 2, 
                               nrows = 1, 
                               figsize=(7,3))

ax1.scatter(x1,y1,
            marker = '*',
            color = "hotpink", 
            edgecolor = "indigo",
           )
ax1.set_xticklabels(ax1.get_xticklabels(), rotation = 45, ha = "right", fontdict=font1)

ax2.bar(x2,y2,
        color = "lightblue",
        edgecolor = "navy")

ax2.set_xticklabels(ax2.get_xticklabels(), rotation = 45, ha = "right", fontdict=font1)



```

```python
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'], 
                                     ['ax2', 'ax3']],
                                    figsize= (7,4),
                                    layout = "constrained")


someaxes['ax1'].scatter(x1,y1)
someaxes['ax2'].bar(x2,y2)
someaxes['ax3'].plot(x1,y1)

someaxes['ax1'].set_xlabel('A Big Label', fontsize = 18)
someaxes['ax2'].set_xlabel("Another Label", fontsize = 18)
someaxes['ax3'].set_xlabel("Lable 2: 2 Fast 2 Furious", fontsize = 18)
```

```python
x = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y1 = np.array([110, 180, 240, 99, 220])
y2 = np.array([170, 100, 90, 120, 50])

fig, ax = plt.subplots(figsize=(7,3))
ax.bar(x,y1, color = "indigo")
ax.plot (x,y2, color = "hotpink")
```

```python
y2_sd = np.std(y2)

fig, ax = plt.subplots(figsize=(7,3))
ax.plot(x, y2, color = "red")

ax.errorbar(x, y2, yerr=y2_sd, fmt = "none", 
            ecolor = "indigo", 
            elinewidth = 4, 
            capsize = 6, 
            capthick = 4
            #errorevery = 2 puts error bar every other 
             ) 
```

```python
from PIL import Image
import requests
from io import BytesIO

response = requests.get("https://static.wikia.nocookie.net/the-ultimate-good/images/3/3d/Luffy.png/revision/latest?cb=20230417200546")
image_file = BytesIO(response.content)
image = Image.open(image_file)
```

```python
fig, ax = plt.subplots(figsize=(7,3))
ax.plot(x, y2, color = "red")

ax_image = fig.add_axes([0.1,0.11, 0.15, 0.35]) #first two numbers are the x and y coordiantor, last two numbers dimensions of the new axes

ax_image.imshow(image)
ax_image.axis("off")

 #removes the axis from the image, so it just looks like an image and not a plot with an image in it
```
