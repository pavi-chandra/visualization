```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests

np.random.seed(613)
```

```python
tips = sns.load_dataset("tips")
print(tips)
```

```python
sns.set_style("darkgrid")

sns.lineplot(data=tips,    #choose our dataset
             x="total_bill", #choose our x variable
             y="tip",)  #choose our y variable
```

```python
fig = plt.subplots(figsize=(10,3))

tipgraph = sns.lineplot(data=tips, 
                        x = "total_bill",
                        y = "tip",
                        color = "hotpink", 
                        linestyle = "--",
                        linewidth = 3,
                        marker = "o",
                        markerfacecolor = "indigo")

tipgraph.set(title = 'Tips vs Total Bill',
            xlabel = 'Total Bill ($)', 
            ylabel = 'Tip Amount ($)'
            )
```

```python
tipgraph = sns.scatterplot(data = tips, 
                           x = 'total_bill',
                           y = 'tip', 
                           style = 'time',  # differentiate points by time of day. the library will pick the preset marker (x, o) for us
                           hue = 'day',     # differentiate points by day of the week
                           palette=['purple', 'hotpink', 'blue', 'green']) # set custom colors for each day

tipgraph.set(title = 'Tips vs Total Bill',
             xlabel = 'Total Bill ($)', 
             ylabel = 'Tip Amount ($)')
```

```python
sns.pairplot(data = tips, hue = 'day') # pairplot a way to quickly visualize relationships. Here we compare day and total bill, hue will assign a colour for each day.  
```

```python
daysplot = sns.relplot (
    data = tips, 
    x = 'total_bill', # x variable
    y = 'tip',        # y variable
    hue = 'sex',      # different colour for each sex
    col = 'day',       # different columnm (graph) for each day
    kind = 'scatter',  # type of graph is scatterplot
    col_wrap = 2        # each row will have two columns of graphs
)
```

```python
import plotly.graph_objects as go
import nbformat 

x1 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y1 = np.array([110, 180, 240, 99, 220])

```

```python
graph =go.Figure()

graph.add_trace(go.Bar(x=x1, y=y1))
graph.update_layout(title = "Pirate Scores",
                    xaxis_title = "Pirates",
                    yaxis_title = "Scores")

graph.show()
```

```python
graph = go.Figure()

graph.add_trace(go.Scatter(x = x1,
                           y = y1,
                           mode = 'markers',  
                           marker = dict(size = 14, 
                                         color = 'hotpink', 
                                         opacity = 1,
                                         line = dict (width = 5, color = 'purple'))))

graph.update_layout(title = "Interactive Pirate Plot",
                    xaxis_title = "Pirates",
                    yaxis_title = "Scores", 
                    width = 500, 
                    height = 500)
```

```python
from wordcloud import WordCloud

#how to import csv from url 

df = pd.read_csv(" https://raw.githubusercontent.com/prasertcbs/basic-dataset/master/movie_quotes.csv",  
                 on_bad_lines = 'skip')
df
```

```python
text = " ".join(each for each in df.quote) # join all the text in quote together into one string, with a space in between each quote

wordcloud = WordCloud(background_color = "white",
                      colormap = 'inferno',
                      random_state = 613).generate(text) # for reproduciablity, so generate the wordcloud with the same colour scheme

fig, ax = plt.subplots(figsize=(7, 3))
ax.imshow(wordcloud)
ax.axis("off")
```

```python
from matplotlib_venn import venn2, venn2_circles, venn2_unweighted

#need to define our sets

A = set(["apple", "banana", "watermelon"])
B = set(["pumpkin", "blueberry", "apple", "key lime"])

diagram = venn2([A, B], 
                set_labels = ("Fruits", "Pies"),
                set_colors = ("blue", "red"), 
                alpha = 0.5)

diagram.get_label_by_id("10").set_text("\n".join(A - B)) # set the text left. Thik of turning off(0) or on (1) so 10 -left is on, but right is off
diagram.get_label_by_id("11").set_text("\n".join(B & A)) # set the text middle
diagram.get_label_by_id("01").set_text("\n".join(B - A)) # set the text right
```
