```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import plotly.graph_objects as go
from  plotly.subplots import make_subplots

np.random.seed(613)
```

```python
#import Wages by education level data sheet from Ontario government website
df = pd.read_csv("https://data.ontario.ca/dataset/1f14addd-e4fc-4a07-9982-ad98db07ef86/resource/4cc07c1b-62ed-4ece-a2a4-d05d0f45081c/download/img-wage-rate-by-edu-age-sex-ft-pt-ca-on-2006-24.csv",  
                 on_bad_lines = 'skip')
df 
```

```python
df["YEAR"].unique() #check the unique values in Year column
```

```python
df.info() #check the data types for each column
```

```python
df = df.rename(columns={"YEAR": "Year", " Men": "Men", " Women": "Women"}) #rename column to remove leanding space and all caps
df
```

```python
new_df = df[(df["TYPE OF WORK"] == " Full-time")].copy() #copy into new dataframe with only full time work data
new_df
group_new_df = new_df.groupby("Year")[["Men", "Women"]].mean().round(2).reset_index() #group by year and then calculate the mean wages for Men and Women. Round to 2 decimal places. Set new index
final = group_new_df.assign(Wage_Gap = group_new_df["Men"] - group_new_df["Women"]) #calculate the wage gap by subtracting wage between men and women 
```

```python
final
```

```python

graph = make_subplots(rows=1, cols=1, specs=[[{"secondary_y": True}]]) #create a subplot with secondary y-axis

#add bar graph for women
graph.add_trace(
    go.Bar(name="Women", 
           x=final["Year"], 
           y=final["Women"], 
           marker_color="#ff719d"), 
           secondary_y=False)

#add bar graph for men
graph.add_trace(
    go.Bar(name="Men", 
           x=final["Year"], 
           y=final["Men"], 
           marker_color="#d3e2ff"), 
           secondary_y=False)

#add line graph for wage gap
graph.add_trace(
    go.Scatter(
        x=final["Year"], 
        y=final["Wage_Gap"], 
        mode="lines+markers", 
        name="Wage Gap", 
        line=dict(color="indigo", 
                  width=3)), 
        secondary_y=True)


graph.update_layout(barmode='relative',   #set bar mode to relative to show the wage gap between men and women
                    title="Wage Discrepancy Between Men and Women betweem 2006 - 2024 (Full-Time Workers)", #add title to the graph
                    xaxis_title="Year", #set x-axis title
                    xaxis_range=[2005.5, 2024.5], #set x-axis range to show all years clearly
                    yaxis_title="Average Wage Rate ($/hr)", #set y-axis title for the bar graph
                    yaxis2_title="Wage Gap ($/hr)", #set secondary y-axis title for the line graph
                    legend_title="Legend",
                    template="plotly_white")

graph.show()
```
